```r
# Load necessary package
library(readr)

# Read CSV file
df <- read_csv("/Users/shithilshetty/Documents/Projects/ctr-prediction-engagement-modeling/data/social_media_ad_optimization_raw.csv")

# Display first rows
head(df)
```

```r
# Drop engagement_score
df <- df |> dplyr::select(-engagement_score)

# Define X (features) and y (target)
X <- df |> dplyr::select(-user_id, -ad_id, -conversion)
y <- df$conversion

# Define feature groups
numeric_features <- c("age", "impressions", "clicks", "time_spent_on_ad")

categorical_features <- c(
  "gender", "location", "interests",
  "ad_category", "ad_platform",
  "ad_type", "day_of_week", "device_type"
)
```

```r
library(recipes)

preprocessor <- recipe(conversion ~ ., data = df) %>%

  # Normalize numeric features (StandardScaler)
  step_normalize(all_of(numeric_features)) %>%

  # One-hot encode categorical variables (OneHotEncoder)
  step_dummy(all_of(categorical_features), one_hot = TRUE)
```

```r
library(tidymodels)

# Define logistic regression model
clf_lr <- logistic_reg(mode = "classification") %>%
  set_engine("glm", maxit = 300)

# Combine preprocessing and model into a workflow
workflow_lr <- workflow() %>%
  add_recipe(preprocessor) %>%
  add_model(clf_lr)
```

```r
library(dplyr)
library(scales)

set.seed(42)

# --- Step 0: Convert target to factor ---
df <- df %>%
  mutate(conversion = as.factor(conversion))

# --- Step 1: Split data ---
data_split <- initial_split(df, prop = 0.8, strata = conversion)
train_data <- training(data_split)
test_data  <- testing(data_split)

# --- Step 2: Prepare predictors ---
predictor_vars <- setdiff(names(df), c("conversion", "user_id", "ad_id"))

# --- Step 3: Fit logistic regression using model.matrix() ---
# model.matrix automatically handles one-hot encoding
formula_lr <- as.formula(paste("conversion ~", paste(predictor_vars, collapse = "+")))

# Fit model on training data
X_train <- model.matrix(formula_lr, data = train_data)[, -1]  # remove intercept column
y_train <- train_data$conversion

glm_fit <- glm(y_train ~ X_train, family = binomial())

# --- Step 4: Prepare full dataset for scoring ---
X_full <- model.matrix(formula_lr, data = df)[, -1]  # remove intercept column

# --- Step 5: Compute linear score ---
coef_vec <- coef(glm_fit)[-1]  # exclude intercept
intercept <- coef(glm_fit)[1]

linear_score <- as.vector(X_full %*% coef_vec + intercept)

# --- Step 6: Scale 0–1 ---
df$engagement_score_supervised <- scales::rescale(linear_score, to = c(0,1))

# --- Step 7: Check results ---
df %>%
  select(user_id, clicks, time_spent_on_ad, ad_platform, ad_category, engagement_score_supervised) %>%
  head()
```

```r
library(yardstick)
library(dplyr)

# --- Step 1: Predict on test data ---
preds <- predict(fit_lr, new_data = test_data, type = "class") %>% pull(.pred_class)

# --- Step 2: Make sure factors have same levels ---
y_true <- factor(test_data$conversion, levels = levels(test_data$conversion))
y_pred <- factor(preds, levels = levels(test_data$conversion))

results <- tibble(truth = y_true, estimate = y_pred)

# --- Step 3: Compute metrics using yardstick explicitly ---
# Accuracy
acc <- yardstick::accuracy(results, truth = truth, estimate = estimate)
print(acc)

# Precision
prec <- yardstick::precision(results, truth = truth, estimate = estimate)
print(prec)

# Recall
rec <- yardstick::recall(results, truth = truth, estimate = estimate)
print(rec)

# F1-score
f1 <- yardstick::f_meas(results, truth = truth, estimate = estimate)
print(f1)
```

```r
library(yardstick)
library(ggplot2)
library(dplyr)
library(tidyr)

# Ensure predictions and truth are factors with same levels
y_true <- factor(test_data$conversion, levels = levels(test_data$conversion))
y_pred <- factor(preds, levels = levels(test_data$conversion))

results <- tibble(truth = y_true, estimate = y_pred)

# Compute confusion matrix
cm_tbl <- conf_mat(results, truth = truth, estimate = estimate)

# Print confusion matrix table
print(cm_tbl)

# Plot confusion matrix
autoplot(cm_tbl, type = "heatmap") +
  scale_fill_gradient(low = "white", high = "blue") +
  labs(title = "Confusion Matrix – Logistic Regression (Conversion Prediction)",
       x = "Predicted", y = "Actual") +
  theme_minimal()
```

```r
library(pROC)
library(ggplot2)
library(dplyr)

# --- Ensure predictions and true labels ---
y_true <- as.numeric(as.character(test_data$conversion))  # convert factor to numeric 0/1
y_prob <- predict(fit_lr, new_data = test_data, type = "prob") %>% pull(.pred_1)

# --- Compute ROC ---
roc_obj <- roc(response = y_true, predictor = y_prob)

# --- AUC value ---
auc_value <- auc(roc_obj)
print(paste("AUC:", round(auc_value, 3)))

# --- Prepare data for ggplot ---
roc_df <- data.frame(
  fpr = 1 - roc_obj$specificities,
  tpr = roc_obj$sensitivities
)

# --- Plot ROC curve ---
ggplot(roc_df, aes(x = fpr, y = tpr)) +
  geom_line(color = "blue", size = 1.2) +
  geom_abline(intercept = 0, slope = 1, linetype = "dashed", color = "gray") +
  labs(
    title = "ROC Curve – Logistic Regression",
    x = "False Positive Rate",
    y = "True Positive Rate",
    subtitle = paste0("AUC = ", round(auc_value, 3))
  ) +
  theme_minimal()
```

```r
library(dplyr)

df <- df %>%
  mutate(
    CTR = if_else(impressions == 0, 0, clicks / impressions)
  )
```

```r
# Ensure target is factor for classification
df <- df %>%
  mutate(conversion = as.factor(conversion))

# ------------------------
# Model 1: CTR-only model
# ------------------------
model_ctr <- glm(conversion ~ CTR, data = df, family = binomial)
# Predicted probabilities
df$prob_ctr <- predict(model_ctr, type = "response")

# ------------------------
# Model 2: Engagement Score–only model
# ------------------------
model_eng <- glm(conversion ~ engagement_score_supervised, data = df, family = binomial)
# Predicted probabilities
df$prob_eng <- predict(model_eng, type = "response")
```

```r
library(pROC)

# Ensure target is numeric for ROC calculation
y <- as.numeric(as.character(df$conversion))

# AUC for CTR-only model
auc_ctr <- roc(y, df$prob_ctr)$auc
cat("AUC - CTR only:", round(auc_ctr, 4), "\n")

# AUC for Engagement Score-only model
auc_eng <- roc(y, df$prob_eng)$auc
cat("AUC - Engagement Score only:", round(auc_eng, 4), "\n")
```
